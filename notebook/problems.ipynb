## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO
import numpy as np


def roll_the_dices(rolltimes=1000):
    counter = 0
    for i in range(rolltimes): 
       dice1 = np.random.randint(1,7)
       dice2 = np.random.randint(1,7) 
       dices_results = dice1 + dice2
       if (dices_results > 7) or (dices_results%2 == 0) :
        counter += 1
    return counter / rolltimes

print(roll_the_dices())
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
import math
### Ejercicio 2
odds_white = 1/6 #prob de sacar blancas (luego rojas y verdes)
odds_red = 1/3
odds_green = 1/2
trials = 5 #numero de ensayos

### Ejercicio 2.1.
#formula probabilidad multinomia -->> p=(n!/(n1!n2!n3!))*((p1**n1)*(p2**m2)*(p3**n3))

# 3 bolitas blancas y 2 bolitas rojas
white_wanted = 3
red_wanted = 2

#coeficiente multinomial y probabilidades totales
multi_coef_2_1 = math.factorial(trials)/ math.factorial(white_wanted) * math.factorial(red_wanted)
total_odds_2_1 = (odds_white**white_wanted) * (odds_red**red_wanted)
final_odds_2_1_exercise = multi_coef_2_1 * total_odds_2_1

print(f'La probabilidad matematica del Ejercicio 2.1 es {final_odds_2_1_exercise:.3%}.')
#---------------------------------------

### Ejercicio 2.2.

all_whites = odds_white ** trials
all_reds = odds_red ** trials
all_greens = odds_green ** trials
total_odds_2_2 = all_whites + all_reds + all_greens
print(f'La probabilidad matematica del Ejercicio 2.2 es {total_odds_2_2:.3%}.')
#------------------------------------------



simulations = 1000
picks = 5

def picking_balls(npicks, n_sim):
    count_2_1 = 0
    count_2_2 = 0
    for i in range(n_sim):
        colorslist = []
        for i in range(npicks):
            colorslist.append(ball_box[np.random.randint(0,59)])
        colorslist = np.array(colorslist)
        whites = sum(colorslist == "White")
        reds = sum(colorslist == "Red")
        greens = sum(colorslist == "Green")
        count_2_1 += 1 if (whites == 3) and (reds == 2) else 0
        count_2_2 +=1 if (whites == 5) or (reds == 5) or (greens == 5) else 0
        
    return count_2_1 / simulations, count_2_2 / simulations

print(picking_balls(picks,simulations))
```
